## Define event windows

### Subtask:
Determine the event date for the cybersecurity event for each company and define the event window and estimation window around this date.

**Reasoning**:
I will create a dictionary of event dates for each company. Then, I will define the event window (e.g., a few days before and after the event) and the estimation window (a period before the event window to estimate normal returns).

```python
from datetime import timedelta

# Define event dates for each company (already defined in previous step)
# event_dates = {
#     'TEAM': pd.to_datetime('2024-01-23'),
#     'ME': pd.to_datetime('2023-11-15'),
#     'INFY': pd.to_datetime('2023-11-10'),
#     'SONY': pd.to_datetime('2023-10-15'),
#     'BA': pd.to_datetime('2023-11-12'),
#     'S63.SI': pd.to_datetime('2020-06-10')
# }

# Define event window and estimation window durations
event_window_days = 5  # e.g., 2 days before and 2 days after the event, plus the event day
estimation_window_days = 100 # e.g., 100 trading days before the event window
post_event_window_days = 30 # Define a post-event window of 30 days

# Calculate event and estimation window dates for each company
event_windows = {}
estimation_windows = {}
post_event_windows = {} # Define post-event windows

for company, event_date in event_dates.items():
    event_windows[company] = (event_date - timedelta(days=event_window_days//2),
                              event_date + timedelta(days=event_window_days//2))
    estimation_windows[company] = (event_date - timedelta(days=event_window_days//2) - timedelta(days=estimation_window_days),
                                   event_date - timedelta(days=event_window_days//2) - timedelta(days=1))
    post_event_windows[company] = (event_date + timedelta(days=event_window_days//2) + timedelta(days=1),
                                   event_date + timedelta(days=event_window_days//2) + timedelta(days=post_event_window_days))


# Display the defined windows
for company in event_dates.keys():
    print(f"Company: {company}")
    print(f"  Event Date: {event_dates[company].strftime('%Y-%m-%d')}")
    print(f"  Event Window: {event_windows[company][0].strftime('%Y-%m-%d')} to {event_windows[company][1].strftime('%Y-%m-%d')}")
    print(f"  Estimation Window: {estimation_windows[company][0].strftime('%Y-%m-%d')} to {estimation_windows[company][1].strftime('%Y-%m-%d')}")
    print(f"  Post-Event Window: {post_event_windows[company][0].strftime('%Y-%m-%d')} to {post_event_windows[company][1].strftime('%Y-%m-%d')}") # Display post-event window
    print("-" * 30)
```

## Calculate daily returns

### Subtask:
Calculate the daily returns for each stock.

**Reasoning**:
To calculate the daily returns, I will iterate through the dataframes, calculate the percentage change in the 'Adj Close' price, and store the returns in a new column or a separate structure. I will use 'Adj Close' as it accounts for dividends and stock splits.

```python
import numpy as np
daily_returns = {}

for sheet_name, df in dataframes.items():
    if df is not None and not df.empty:
        # Ensure the 'Date' column is datetime and set as index for easier slicing
        df['Date'] = pd.to_datetime(df['Date'])
        df = df.set_index('Date')
        # Calculate daily returns using 'Adj Close'
        # Handle potential division by zero if 'Adj Close' is 0
        daily_returns[sheet_name] = df['Adj Close'].pct_change().replace([np.inf, -np.inf], np.nan)
        print(f"Daily Returns calculated for {sheet_name}.")
        # Display the first few rows of the returns
        display(daily_returns[sheet_name].head())
        print("\n")
    else:
        print(f"DataFrame for {sheet_name} is not valid for calculating returns.")
```

## Estimate expected returns

### Subtask:
For each stock, estimate the expected return during the event window using a market model or similar approach, based on data from the estimation window.

**Reasoning**:
To estimate expected returns, I will use a market model. This involves regressing each stock's daily returns on the returns of a market index during the estimation window. Since a market index is not provided, I will use the average return of all other stocks in the dataset as a proxy for the market return for each individual stock's estimation. This is a simplification, and using a relevant market index would be more accurate if available.

```python
import statsmodels.api as sm

expected_returns = {}
market_model_results = {}

for company, returns in daily_returns.items():
    if returns is not None and not returns.empty:
        # Filter returns for the estimation window
        estimation_returns = returns.loc[estimation_windows[company][0]:estimation_windows[company][1]].dropna()

        # Create a proxy for market return using the average of other stocks
        other_stocks = [name for name in daily_returns.keys() if name != company]
        if other_stocks:
            # Combine and average returns of other stocks
            market_proxy_returns = pd.concat([daily_returns[other].loc[estimation_windows[company][0]:estimation_windows[company][1]] for other in other_stocks], axis=1).mean(axis=1).dropna()

            # Ensure both series cover the same time period for regression
            common_index = estimation_returns.index.intersection(market_proxy_returns.index)
            estimation_returns_aligned = estimation_returns.loc[common_index]
            market_proxy_returns_aligned = market_proxy_returns.loc[common_index]


            if not estimation_returns_aligned.empty and not market_proxy_returns_aligned.empty:
                # Add a constant for the regression
                X = sm.add_constant(market_proxy_returns_aligned)
                y = estimation_returns_aligned

                # Fit the market model
                model = sm.OLS(y, X).fit()
                market_model_results[company] = model
                print(f"Market model fitted for {company}.")
                # print(model.summary()) # Uncomment to see regression results

                # Predict expected returns during the event window
                event_returns = returns.loc[event_windows[company][0]:event_windows[company][1]].dropna()
                # Align market proxy returns for the event window
                market_proxy_event_returns = pd.concat([daily_returns[other].loc[event_windows[company][0]:event_windows[company][1]] for other in other_stocks], axis=1).mean(axis=1).dropna()

                # Ensure both series cover the same time period for prediction
                event_common_index = event_returns.index.intersection(market_proxy_event_returns.index)
                market_proxy_event_returns_aligned = market_proxy_event_returns.loc[event_common_index]

                if not market_proxy_event_returns_aligned.empty:
                    X_event = sm.add_constant(market_proxy_event_returns_aligned)
                    expected_returns[company] = model.predict(X_event)
                    print(f"Expected returns calculated for {company} during the event window.")
                else:
                    print(f"Not enough data in the event window to predict expected returns for {company}.")

            else:
                print(f"Not enough overlapping data in the estimation window for {company} and market proxy to fit the model.")
        else:
            print(f"No other stocks available to create a market proxy for {company}.")
    else:
        print(f"Daily returns not available or empty for {company}.")

# Display first few expected returns for each company
print("\n--- Expected Returns (first 5 rows) ---")
for company, exp_ret in expected_returns.items():
    print(f"Expected Returns for {company}:")
    display(exp_ret.head())
    print("-" * 30)
```

## Calculate abnormal returns

### Subtask:
Calculate the abnormal return for each stock during the event window by subtracting the expected return from the actual return.

**Reasoning**:
Abnormal returns are calculated as the difference between the actual daily returns and the expected returns estimated by the market model. I will iterate through the companies for which we have expected returns and calculate the abnormal returns during their event windows.

```python
abnormal_returns = {}

for company, exp_ret in expected_returns.items():
    if company in daily_returns and daily_returns[company] is not None:
        # Calculate abnormal returns during the event window
        actual_returns_event = daily_returns[company].loc[event_windows[company][0]:event_windows[company][1]].dropna()

        # Align actual and expected returns for the event window
        common_index_event = actual_returns_event.index.intersection(exp_ret.index)
        actual_returns_event_aligned = actual_returns_event.loc[common_index_event]
        exp_ret_event_aligned = exp_ret.loc[common_index_event]

        if not actual_returns_event_aligned.empty and not exp_ret_event_aligned.empty:
            abnormal_returns[company] = actual_returns_event_aligned - exp_ret_event_aligned
            print(f"Abnormal returns calculated for {company} during the event window.")
            # display(abnormal_returns[company].head()) # Display head for the event window abnormal returns
            # print("\n")
        else:
            print(f"Not enough overlapping data in the event window to calculate abnormal returns for {company}.")

        # Calculate abnormal returns during the post-event window
        actual_returns_post_event = daily_returns[company].loc[post_event_windows[company][0]:post_event_windows[company][1]].dropna()

        # Align actual and expected returns for the post-event window
        # Note: We use the same market model from the estimation window to predict expected returns
        # in the post-event window.
        market_proxy_post_event_returns = pd.concat([daily_returns[other].loc[post_event_windows[company][0]:post_event_windows[company][1]] for other in daily_returns.keys() if other != company], axis=1).mean(axis=1).dropna()

        common_index_post_event = actual_returns_post_event.index.intersection(market_proxy_post_event_returns.index)
        actual_returns_post_event_aligned = actual_returns_post_event.loc[common_index_post_event]
        market_proxy_post_event_returns_aligned = market_proxy_post_event_returns.loc[common_index_post_event]

        if not actual_returns_post_event_aligned.empty and not market_proxy_post_event_returns_aligned.empty and company in market_model_results:
             X_post_event = sm.add_constant(market_proxy_post_event_returns_aligned)
             expected_returns_post_event = market_model_results[company].predict(X_post_event)
             # Ensure alignment between actual and predicted returns in the post-event window
             post_event_common_index = actual_returns_post_event_aligned.index.intersection(expected_returns_post_event.index)
             actual_returns_post_event_final = actual_returns_post_event_aligned.loc[post_event_common_index]
             expected_returns_post_event_final = expected_returns_post_event.loc[post_event_common_index]

             if not actual_returns_post_event_final.empty and not expected_returns_post_event_final.empty:
                abnormal_returns[company] = pd.concat([abnormal_returns[company], (actual_returns_post_event_final - expected_returns_post_event_final)])
                print(f"Abnormal returns calculated for {company} during the post-event window.")
                display(abnormal_returns[company].tail()) # Display tail for the combined abnormal returns
                print("\n")
             else:
                 print(f"Not enough overlapping data in the post-event window to calculate abnormal returns for {company}.")

        else:
            print(f"Not enough data or market model not available for {company} in the post-event window.")

    else:
        print(f"Daily returns not available for {company} to calculate abnormal returns.")

# Note: Abnormal returns for S63.SI cannot be calculated as expected returns were not estimated.
```

## Calculate cumulative abnormal returns

### Subtask:
Calculate the cumulative abnormal return (CAR) for each stock over the event window.

**Reasoning**:
The cumulative abnormal return (CAR) is the sum of the abnormal returns over the event window. I will calculate the CAR for each company for which abnormal returns were calculated.

```python
cumulative_abnormal_returns = {}

for company, ab_ret in abnormal_returns.items():
    if ab_ret is not None and not ab_ret.empty:
        cumulative_abnormal_returns[company] = ab_ret.sum()
        print(f"Cumulative Abnormal Return (CAR) for {company} (including post-event window): {cumulative_abnormal_returns[company]:.4f}")
    else:
        print(f"No abnormal returns available for {company} to calculate CAR.")

# Note: CAR for S63.SI cannot be calculated as abnormal returns were not calculated.
```

## Perform statistical tests

### Subtask:
Conduct statistical tests to determine if the CARs are statistically significant.

**Reasoning**:
I will perform a t-test on the abnormal returns for each company over their respective event windows to assess the statistical significance of the cumulative abnormal returns.

```python
from scipy import stats

print("--- Statistical Significance of CARs (including post-event window) ---")

statistical_results = {}

for company, ab_ret in abnormal_returns.items():
    if ab_ret is not None and not ab_ret.empty:
        # Perform a t-test on the abnormal returns over the event and post-event window
        # Null hypothesis: The mean abnormal return during the combined window is zero.
        # Alternative hypothesis: The mean abnormal return during the combined window is not zero.
        t_statistic, p_value = stats.ttest_1samp(ab_ret, 0)

        statistical_results[company] = {
            't_statistic': t_statistic,
            'p_value': p_value
        }

        print(f"\nCompany: {company}")
        print(f"  T-statistic: {t_statistic:.4f}")
        print(f"  P-value: {p_value:.4f}")

        # Interpret the p-value (common significance level is 0.05)
        alpha = 0.05
        if p_value < alpha:
            print(f"  The CAR for {company} (including post-event window) is statistically significant at the {alpha} level.")
        else:
            print(f"  The CAR for {company} (including post-event window) is not statistically significant at the {alpha} level.")
    else:
        print(f"\nNo abnormal returns available for {company} to perform statistical testing.")

# Note: Statistical test for S63.SI cannot be performed as abnormal returns were not calculated.
```

## Estimate expected returns (Including S63.SI)

### Subtask:
For each stock, estimate the expected return during the event window using a market model or similar approach, based on data from the estimation window, including S63.SI.

**Reasoning**:
I will re-run the market model estimation for all companies, including S63.SI, now that the data for S63.SI has been confirmed to cover the necessary period.

```python
import statsmodels.api as sm

expected_returns = {}
market_model_results = {}

for company, returns in daily_returns.items():
    if returns is not None and not returns.empty:
        # Filter returns for the estimation window
        estimation_returns = returns.loc[estimation_windows[company][0]:estimation_windows[company][1]].dropna()

        # Create a proxy for market return using the average of other stocks
        other_stocks = [name for name in daily_returns.keys() if name != company]
        if other_stocks:
            # Combine and average returns of other stocks
            market_proxy_returns = pd.concat([daily_returns[other].loc[estimation_windows[company][0]:estimation_windows[company][1]] for other in other_stocks], axis=1).mean(axis=1).dropna()

            # Ensure both series cover the same time period for regression
            common_index = estimation_returns.index.intersection(market_proxy_returns.index)
            estimation_returns_aligned = estimation_returns.loc[common_index]
            market_proxy_returns_aligned = market_proxy_returns.loc[common_index]


            if not estimation_returns_aligned.empty and not market_proxy_returns_aligned.empty:
                # Add a constant for the regression
                X = sm.add_constant(market_proxy_returns_aligned)
                y = estimation_returns_aligned

                # Fit the market model
                model = sm.OLS(y, X).fit()
                market_model_results[company] = model
                print(f"Market model fitted for {company}.")
                # print(model.summary()) # Uncomment to see regression results

                # Predict expected returns during the event window
                event_returns = returns.loc[event_windows[company][0]:event_windows[company][1]].dropna()
                # Align market proxy returns for the event window
                market_proxy_event_returns = pd.concat([daily_returns[other].loc[event_windows[company][0]:event_windows[company][1]] for other in other_stocks], axis=1).mean(axis=1).dropna()

                # Ensure both series cover the same time period for prediction
                event_common_index = event_returns.index.intersection(market_proxy_event_returns.index)
                market_proxy_event_returns_aligned = market_proxy_event_returns.loc[event_common_index]

                if not market_proxy_event_returns_aligned.empty:
                    X_event = sm.add_constant(market_proxy_event_returns_aligned)
                    expected_returns[company] = model.predict(X_event)
                    print(f"Expected returns calculated for {company} during the event window.")
                else:
                    print(f"Not enough data in the event window to predict expected returns for {company}.")

            else:
                print(f"Not enough overlapping data in the estimation window for {company} and market proxy to fit the model.")
        else:
            print(f"No other stocks available to create a market proxy for {company}.")
    else:
        print(f"Daily returns not available or empty for {company}.")

# Display first few expected returns for each company
print("\n--- Expected Returns (first 5 rows) ---")
for company, exp_ret in expected_returns.items():
    print(f"Expected Returns for {company}:")
    display(exp_ret.head())
    print("-" * 30)
```
